# Modelo KNN (K-Nearest Neighbor)
Nesse notebook, foi feito a implementação do modelo de predição KNN, onde a máquina faz uma classificação de dados baseado nos dados classificado mais próximos.
Entre os outros modelos, KNN é o modelo mais apropriado para classificações, mas pode ser usado sim para outros meios.

```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model, preprocessing
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Depois de ter feito a importação das bibliotecas, mostrei as últimas cinco entradas da tabela para ver se tinha algo de errado.

```python
dados = pd.read_csv('../content/drive/MyDrive/Planilhas/car.csv')
dados.tail()
```

Aqui foi feito o uso do preprocessing para transformar os dados nas colunas que tem texto em números, já que a máquina precisaria dos dados como números inteiros para trabalhar adequadamente.

```python
le = preprocessing.LabelEncoder()

comprando = le.fit_transform(list(dados['buying']))
manut = le.fit_transform(list(dados['maint']))
porta = le.fit_transform(list(dados['door']))
pessoas = le.fit_transform(list(dados['persons']))
porta_mala = le.fit_transform(list(dados['lug_boot']))
segurança = le.fit_transform(list(dados['safety']))
classe = le.fit_transform(list(dados['class']))
```

Aqui fiz todo o procedimento padrão para machine learning e treinei a máquina usando 20% dos dados, porém eu mudei o modelo para KNeighborsClassifier.
Fiz uma lista com os nomes das classes para mudar de numeros para nomes novamente.
Esse modelo tem entre 88% e 95% de taxa de precisão nas previsões que tem.

```python
X = list(zip(comprando, manut, porta, pessoas, porta_mala, segurança))
y = list(classe)

x_treino, x_teste, y_treino, y_teste = sklearn.model_selection.train_test_split(X, y, test_size=0.2)

modelo = KNeighborsClassifier(n_neighbors=7)

modelo.fit(x_treino, y_treino)

prec = modelo.score(x_teste, y_teste)

print(prec)

previsto = modelo.predict(x_teste)

nomes = ['0 - unacc', '1 - acc',  '2 - good', '3 - vgood']

for x in range(len(previsto)):
  print(f'Previsto: {nomes[previsto[x]]}, Dados: {x_teste[x]}, Resultado final:{nomes[y_teste[x]]} ')
  print()
```
